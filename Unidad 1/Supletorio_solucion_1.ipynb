```python
# Clase base
class Organismo:
    def __init__(self, nombre, edad, especie):
        self.nombre = nombre
        self.edad = edad
        self.especie = especie
    
    def respirar(self):
        print(f"{self.nombre} está respirando")
    
    def crecer(self):
        self.edad += 1
        print(f"{self.nombre} ha crecido. Ahora tiene {self.edad} años.")

# Clase hija que hereda de Organismo
class Planta(Organismo):
    def __init__(self, nombre, edad, especie, tipo):
        super().__init__(nombre, edad, especie)
        self.tipo = tipo
    
    def fotosintesis(self):
        print(f"{self.nombre} está haciendo fotosíntesis")
    
    def crecer(self):
        self.edad += 0.5
        print(f"{self.nombre} ha crecido. Ahora tiene {self.edad} años.")

# Clase hija que hereda de Organismo
class Animal(Organismo):
    def __init__(self, nombre, edad, especie, tipo_alimentacion):
        super().__init__(nombre, edad, especie)
        self.tipo_alimentacion = tipo_alimentacion
    
    def moverse(self):
        print(f"{self.nombre} está moviéndose")
    
    def crecer(self):
        self.edad += 2
        print(f"{self.nombre} ha crecido. Ahora tiene {self.edad} años.")

# Ejemplo de uso
planta1 = Planta("Rosa", 1, "Rosaceae", "Ornamental")
planta1.fotosintesis()
planta1.crecer()

animal1 = Animal("León", 5, "Panthera leo", "Carnívoro")
animal1.moverse()
animal1.crecer()
```
